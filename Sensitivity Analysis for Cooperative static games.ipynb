```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
a_values = np.linspace(0, 1, 100)  # Values for parameter 'a'
b_values = np.linspace(0, 1, 100)  # Values for parameter 'b'

# Define a function to calculate the outcome variable
def calculate_outcome(a, b):
    return a * b  # Example mathematical relationship, can be replaced with actual function

# Perform sensitivity analysis for parameter 'a'
outcome_values_a = [calculate_outcome(a, 41.3) for a in a_values]

# Perform sensitivity analysis for parameter 'b'
outcome_values_b = [calculate_outcome(0.5, b) for b in b_values]

# Plotting for parameter 'a'
plt.figure(figsize=(8, 6))
plt.plot(a_values, outcome_values_a)
plt.xlabel('Parameter a')
plt.ylabel('Outcome')
plt.title('Sensitivity Analysis for Parameter a')
plt.grid(True)
plt.show()

# Plotting for parameter 'b'
plt.figure(figsize=(8, 6))
plt.plot(b_values, outcome_values_b)
plt.xlabel('Parameter b')
plt.ylabel('Outcome')
plt.title('Sensitivity Analysis for Parameter b')
plt.grid(True)
plt.show()
```
